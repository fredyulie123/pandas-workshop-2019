## How to filter rows of a DataFrame by columns?

```python
#check shape of dataset
movies.shape
```

```python
#find movies that are really long (at least 200 minutes)
booleans = []
for length in movies.duration:    
    if length >= 200:        
        booleans.append(True)    
    else:        
        booleans.append(False)

booleans[0:5]
```

```python
is_long=pd.Series(booleans)
is_long.head()
```

```python
movies[is_long]
```

```python
df=pd.DataFrame[is_long]
```

```python
# 2 Using Pandas DataFrame Method
is_long=movies.duration >= 200
is_long.head()
movies[is_long]
```

```python
# 3
movies[movies.duration >= 200]
```

```python
#only see the “genre” of movies duration >=200
movies[movies.duration >= 200]['genre']
#movies[movies.duration >= 200].genre
#movies.loc[movies.duration >= 200,'genre']
```

```python
#filter data
#only look at long movies also the genre of the drama by Using "and"
movies[(movies.duration >=200) & (movies.genre =='Drama')]
```

```python
#look at long movies or genre of the drama by Using "or" "|"-->pipe character
movies[(movies.duration >=200) | (movies.genre =='Drama')]
```

```python
#What if you have a bunch of conditions that movies are crime, action, or drama?
movies[(movies.genre =='Drama') | (movies.genre =='Crime') | (movies.genre =='Action')]
```

```python
#Alternatively, you can use a Series method:
movies[movies.genre.isin(['Crime','Drama','Action'])]
```

## How to use Groupby in Pandas?

```python
drinks = pd.read_csv('drinks.csv')
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
#For each continent, what is the average beer_servings?
drinks.groupby('continent').beer_servings.mean()
```

```python
#If you only want to see the average beer servings in Africa:
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
#You can also use “max”, “min” as aggregate functions:
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.min()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```python
drinks.groupby('continent').mean()
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```

## How to handle missing values?

```python
#handle missing values
ufo = pd.read_csv('ufo.csv')
ufo.tail()
```

```python
#check the missing values by using isnull() function:
ufo.isnull().tail()
```

```python
#You can check missing values by columns
ufo.isnull().sum()
```

```python
ufo[ufo.City.isnull()]
```

```python
#Check the shape of the data
ufo.shape 
```

```python
#Drop the missing values
# If the row has one 'na', we will drop the row
ufo.dropna(how='any').shape
```

```python
# If the row all the indexes are 'na', we will drop it
ufo.dropna(how='all').shape
```

```python
#If you want to drop missing values by specific requirements, for example
ufo.dropna(subset=['City','Shape Reported'],how='any').shape
```

```python
#Filing the missing values: 
ufo['Shape Reported'].value_counts(dropna=False)
```

```python
ufo['Shape Reported'].fillna(value='IRREGULAR',inplace=True)
ufo['Shape Reported'].value_counts(dropna=False)
```

### outer/ inner/ left/ right merge 

```python
df1 = pd.DataFrame({'Year':[2001,2002,2003,2004],
'Int_rate':[2, 3, 2, 2],
'US_GDP_Thousands':[50, 55, 65, 55]})

df3 = pd.DataFrame({'Year':[2001,2003,2004,2005],
'Unemployment':[7, 8, 9, 6],
'Low_tier_HPI':[50, 52, 50, 53]})
```

```python
df1
```

```python
df3
```

```python
inner_merged = pd.merge(df1, df3,on = 'Year', how = 'inner') 
inner_merged
```

```python
outer_merged = pd.merge(df1, df3,on = 'Year', how = 'outer') 
outer_merged
```

```python
# left_merged is choose df1 as standard
left_merged = pd.merge(df1, df3,on = 'Year', how = 'left') 
left_merged
```

```python
# right_merged is choose df2 as standard
right_merged = pd.merge(df1, df3,on = 'Year', how = 'right')
right_merged
```
